```python
from transformers import pipeline
import gradio as gr

# Load GPT-2 model for text generation
generator = pipeline("text-generation", model="gpt2")

# Genre templates
genres = {
    "Sci-Fi": "In a distant future, ",
    "Horror": "It was a dark and stormy night. ",
    "Romance": "In the heart of Paris, ",
    "Fantasy": "Once upon a time in a magical realm, ",
    "Mystery": "The detective walked into the silent room. "
}

# Story generator function
def generate_story(prompt, genre, temperature, length):
    full_prompt = genres[genre] + prompt
    output = generator(full_prompt, max_length=int(length), temperature=float(temperature), num_return_sequences=1)
    return output[0]['generated_text']

# Create Gradio interface
demo = gr.Interface(
    fn=generate_story,
    inputs=[
        gr.Textbox(label="Your Prompt", placeholder="e.g., a robot learned how to love"),
        gr.Dropdown(list(genres.keys()), label="Genre"),
        gr.Slider(0.7, 1.3, value=1.0, label="Creativity (temperature)"),
        gr.Slider(50, 300, value=150, step=10, label="Story Length (tokens)")
    ],
    outputs=gr.Textbox(label="Generated Story"),
    title="📝 GPT-2 Story Generator",
    description="Enter a story prompt, choose a genre, and generate a creative story using GPT-2 for free!",
)

demo.launch(share=True)  # Gives a public link
```
